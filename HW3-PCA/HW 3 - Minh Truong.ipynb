**Import libraries**

```python
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
import matplotlib.cm as cm
```

**1. Load the dataset**

```python
# Load data
iris = load_iris()
data = iris.data
```

**Understand the data**

```python
n_samples, n_features = data.shape
n_iris = len(np.unique(iris.target))
labels = iris.target
```

```python
print(f"Number of samples: {n_samples}")
print(f"Number of features: {n_features}")
print(f"Number of iris species: {n_iris}")
print(f"Labels array shape: {labels.shape}")
print(f"Unique labels: {np.unique(labels)}")
print(f"Label distribution: {np.bincount(labels)}") 
```

**2. Standardize the Data**

```python
# Standardize the data (recommended before PCA)
data_scaled = scale(data)
```

**3. Apply PCA from scikit-learn/Initialize PCA algorithm and apply it to the dataset:**

```python
pca = PCA(n_components=2)
data_r = pca.fit_transform(data_scaled)
```

**Examining the results**

```python
print('Explained variance ratio (first two components): %s' %
      str(pca.explained_variance_ratio_))
print('Sum of explained variance (first two components): %s' %
      str(sum(pca.explained_variance_ratio_)))
```

**4. Visualize the Results**

```python
plt.figure(figsize=(10, 6))

# Plot each species separately
for i in range(n_iris):
    mask = labels == i
    plt.scatter(data_r[mask, 0], data_r[mask, 1], 
                label=iris.target_names[i], 
                s=50, alpha=0.7)

# Add variance percentages to axis labels
pc1_var = pca.explained_variance_ratio_[0] * 100
pc2_var = pca.explained_variance_ratio_[1] * 100

plt.xlabel(f'PC1 ({pc1_var:.1f}% variance)')
plt.ylabel(f'PC2 ({pc2_var:.1f}% variance)')
plt.title('PCA of Iris Dataset')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
